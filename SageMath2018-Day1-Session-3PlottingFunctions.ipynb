## Plotting more than one graph together

### Method 1

```python
p1 = plot(sin(x), 
         (x,-pi,pi),
         color = 'red')
p2 = plot(cos(x),
         (x,-pi,pi))
p1 + p2
```

### Method 2

```python
g = Graphics()
g += plot(sin(x), (x,-pi,pi), color="red", legend_label = '$\sin x$',
         title = "Graph of sin x and cos x") # of order 1
g += plot(cos(x), (x,-pi,pi), color ="blue", legend_label = '$\cos x$') # of order 2
g += points((pi/4,sin(pi/4)), pointsize = 50, color = 'green', zorder = 3)
g += circle((pi/4,sin(pi/4)),.15, linestyle = 'dashed', color = 'red')
g += text("SageMath is using Mathplotlib for plotting\n the functions.And its amazing...", 
          (-0.5,-0.5),
         horizontal_alignment = 'left',
         color = 'teal',
         bounding_box={'boxstyle':'round', 'fc':'w'},
         clip = True)
    
g.show(figsize = 9)
```
