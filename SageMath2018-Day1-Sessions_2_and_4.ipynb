### Basic Graph Operations

````python
G1 = graphs.CycleGraph(3)
G2 = graphs.PathGraph(4)

H = G1.disjoint_union(G2).
H.show(vertex_size = 1000, figsize = 10)
````

````python
G1 = graphs.CycleGraph(3)
G2 = graphs.PathGraph(4)

H = G1.cartesian_product(G2)
H.show(vertex_size = 500, figsize = 8)
````

````python
G1 = graphs.CycleGraph(3)
G2 = graphs.PathGraph(4)

H = G1.categorical_product(G2)
H.show(vertex_size = 500, figsize = 8)
````

## Functions

````python
def dbl(x):
    return 2*x

print dbl(102)
````

````python
def sumRange(a, b, p = 1):
    if a > b:
        a, b = b, a
        
    s = 0
    for k in range(a, b+1):
        s += k^p
    return s

print sumRange(10, 1)
````
